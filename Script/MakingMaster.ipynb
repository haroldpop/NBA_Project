```python
def adjust_id_and_concatenate(df1, df2):
    max_id = df1.iloc[-1]['GAMEID']
    new_id = max_id + 1
    df2_copy = df2.copy()
    for k in range(df2.shape[0]):
        df2_copy.loc[k, 'GAMEID'] = df2_copy.loc[k, 'GAMEID'] + new_id
    df_res = pd.concat([df1, df2_copy], axis = 0)
    df_res.reset_index(inplace = True)
    df_res.drop(columns = ['index'], inplace = True)
    return df_res

def following_ids(df_scores, df):
    #make sure the game ID match between the score and a dataframe team_stats
    df_scores_copy = df_scores.copy()
    df_copy = df.copy()
    for i in range(df_scores.shape[0]):
        id_team = df_scores_copy.loc[i, 'Home'].split(' ')[-1]
        game_id = df_scores_copy.loc[i, 'GAMEID']
        #print(game_id, id_team)
        try:
            ob_team = df_copy[df_copy['GAMEID'] == game_id].iloc[-1]['TEAM'].split(' ')[-1]
        except IndexError:
            continue
        if ob_team != id_team:
            print(id_team, ob_team, game_id)
            print('ERROR AT GAME ID {}'.format(game_id))
            print('TEAM DF_TEAM {}'.format(ob_team))
            print('TEAM DF_SCORE {}'.format(id_team))
            return df_scores[i, i+10]
        
def add_date(df_scores, df2):
    df_copy = df2.copy()
    df_copy['Date'] = ''
    df_scores_copy = df_scores.copy()
    for i in range(df2.shape[0]):
        game_id = df_copy.loc[i, 'GAMEID']
        date = df_scores_copy['Date'][df_scores_copy['GAMEID'] == game_id]
        df_copy.loc[i, 'Date'] = date.values
    return df_copy
```

```python
def change_date_format(df_scores):
    df_scores_copy = df_scores.copy()
    for k in range(df_scores_copy.shape[0]):
        year = df_scores.loc[k, 'Date'].split('/')[2]
        day = df_scores.loc[k, 'Date'].split('/')[1]
        month = df_scores.loc[k, 'Date'].split('/')[0]
        df_scores_copy.loc[k, 'Date'] = year + '-' + month + '-' + day
    return df_scores_copy


#change of the date format in score
df_scores_17_good = change_date_format(df_scores_17)
df_scores_18_good = change_date_format(df_scores_18)
df_scores_20_good = change_date_format(df_scores_20)

#adding of the date columns in team stats and players stats DataFrame
df_team_10_good = add_date(df_scores_10, df_team_10)
df_stats_10_good = add_date(df_scores_10, df_stats_10)

df_team_15_good = add_date(df_scores_15, df_team_15)
df_stats_15_good = add_date(df_scores_15, df_stats_15)

df_team_17_good = add_date(df_scores_17_good, df_team_17)
#df_stats_17_good = add_date(df_scores_17_good, df_stats_17)

df_team_18_good = add_date(df_scores_18_good, df_team_18)
#df_stats_18_good = add_date(df_scores_18_good, df_stats_18)

df_team_20_good = add_date(df_scores_20_good, df_team_20)
#df_stats_20_good = add_date(df_scores_20_good, df_stats_20)
```

```python
def add_win_column(df_scores, df):
    df_scores_copy = df_scores.copy()
    df_copy = df.copy()
    df_copy['WIN'] = 0
    for k in range(df_scores_copy.shape[0]):
        game_id = df_scores_copy.loc[k, 'GAMEID']
        if df_scores_copy.loc[k, 'VisitorScore'] < df_scores_copy.loc[k, 'HomeScore']:
            row_l = df_copy.index[(df_copy['HOME'] == 0) & (df_copy['GAMEID'] == game_id)]
            row_w = df_copy.index[(df_copy['HOME'] == 1) & (df_copy['GAMEID'] == game_id)]
        else:
            row_w = df_copy.index[(df_copy['HOME'] == 0) & (df_copy['GAMEID'] == game_id)]
            row_l = df_copy.index[(df_copy['HOME'] == 1) & (df_copy['GAMEID'] == game_id)]
        df_copy.loc[row_w, 'WIN'] = 1
        df_copy.loc[row_l, 'WIN'] = 0        
    return df_copy


df_team_10_w = add_win_column(df_scores_10, df_team_10)
df_team_11_w = add_win_column(df_scores_11, df_team_11)
df_team_12_w = add_win_column(df_scores_12, df_team_12)
df_team_13_w = add_win_column(df_scores_13, df_team_13)
df_team_14_w = add_win_column(df_scores_14, df_team_14)
df_team_15_w = add_win_column(df_scores_15, df_team_15)
df_team_16_w = add_win_column(df_scores_16, df_team_16)
df_team_17_w = add_win_column(df_scores_17, df_team_17)
df_team_18_w = add_win_column(df_scores_18, df_team_18)
df_team_20_w = add_win_column(df_scores_20, df_team_20)
```
