```python
def r_square(yhat, y):
    y_mean = np.mean(y)
    num = np.sum((yhat - y)**2)
    den = np.sum((y - y_mean)**2)
    print(num, den)
    return 1 - (num / den)
```

```python
team = list(df['TEAM'].unique())[0]
df_team = df[df['TEAM'] == team]
df_team
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import seaborn as sns

scaler = StandardScaler()

df_show = df_team.drop(columns = ['TEAM', 'W', 'L', 'WIN%'])
list_attributes = list(df_show.columns)

X = np.array(df_team.drop(columns = ['TEAM', 'W', 'L', 'WIN%']))
y = np.array(df_team['W'])
y = y.reshape(-1, 1)

X = scaler.fit_transform(X)
y = scaler.fit_transform(y)

print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1, shuffle = True)

print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)


reg = LinearRegression()
reg.fit(X_train, y_train)
yhat = reg.predict(X_test)
print(y_test, yhat)
df_plot = pd.DataFrame(reg.coef_, columns = list_attributes)
plt.figure(figsize = (20, 10))
sns.barplot(data = df_plot)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

scaler = StandardScaler()

X = np.array(df_team.drop(columns = ['TEAM', 'W', 'L', 'WIN%']))
y = np.array(df_team['W'])
y = y.reshape(-1, 1)

poly = PolynomialFeatures(degree = 2)
X = poly.fit_transform(X)
X = scaler.fit_transform(X)
y = scaler.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1, shuffle = True)

print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)

reg = LinearRegression(fit_intercept = False) #already done in PolynomialFeatures
reg.fit(X_train, y_train)
yhat = reg.predict(X_test)
print(y_test, yhat)
```
